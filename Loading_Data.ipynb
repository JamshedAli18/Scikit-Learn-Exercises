# Loading Data

```python
# Load the Iris dataset and display the first 5 rows.
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Display the first 5 rows
iris_df.head()
```

```python
# Load the digits dataset and count the number of samples.
from sklearn.datasets import load_digits

# Load digits dataset
digits = load_digits()

# Count the number of samples
num_samples = digits.data.shape[0]
print(f'Number of samples in the digits dataset: {num_samples}')
```

```python
# Load the wine dataset and print the feature names.
from sklearn.datasets import load_wine

# Load wine dataset
wine = load_wine()

# Print feature names
print(f'Feature names in the wine dataset: {wine.feature_names}')
```

```python
# Load the breast cancer dataset and check for missing values.
from sklearn.datasets import load_breast_cancer

# Load breast cancer dataset
breast_cancer = load_breast_cancer()
breast_cancer_df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)

breast_cancer_df.head()
```

```python
# Check for missing values
missing_values = breast_cancer_df.isnull().sum()
print(f'Missing values in the breast cancer dataset:\n{missing_values}')
```
